# Smart Quiz & Performance Analyzer System (Console-Based)

## Objective
This system conducts a simple console-based quiz using only `if-else` conditional statements and basic operators, without relying on loops. It tracks user scores and provides a performance analysis based on the final score.

## Functional Requirements

1.  **Quiz System**:
    *   Presents Multiple Choice Questions (MCQs) to the user.
    *   Accepts user input for answers.
    *   Validates answers using `if-else` statements.

2.  **Score Tracking**:
    *   Initializes `score` to 0.
    *   Increments `score` by 1 for each correct answer.

3.  **Performance Analysis**:
    *   Classifies performance into categories: 'Advanced', 'Intermediate', 'Beginner', and 'Needs improvement'.
    *   Provides feedback based on the total score out of 10 questions.

4.  **Input Handling**:
    *   Converts user input to uppercase for case-insensitive answer checking.
    *   While basic, this fulfills the input handling without loops.

## Technical Implementation Details

*   **No Loops**: The core constraint of this implementation is the strict avoidance of any looping constructs (`for`, `while`). Each question is asked and processed sequentially by explicitly writing out the `print`, `input`, and `if-else` blocks for every question.
*   **Variables**: A `score` variable tracks correct answers.
*   **Conditional Statements**: `if-else` blocks are used extensively to check user answers against correct answers and to determine the performance classification.
*   **Operators**: Basic arithmetic operators (`+`) for scoring and comparison operators (`==`, `>=`, `elif`) for evaluation are used.
*   **Type Casting**: `input().upper()` is used to convert string input to uppercase for standardized comparison.

## Quiz Flow

1.  The quiz starts with a welcome message.
2.  Each question is displayed, followed by its options.
3.  The user provides an answer, which is converted to uppercase.
4.  An `if-else` statement checks if the answer is correct; if so, the score is incremented.
5.  This process is repeated for all 10 questions.
6.  Finally, the total score is displayed, and a performance analysis is given based on the score thresholds.

## Scoring and Performance Thresholds (Out of 10 questions)

*   **Advanced**: Score >= 8
*   **Intermediate**: Score >= 6
*   **Beginner**: Score >= 4
*   **Needs Improvement**: Score < 4

## Simple Quiz System (No Loops, only if-else)

```python
score = 0

print("Welcome to the Simple Python Quiz!")
print("Please answer A, B, C, or D for each question.")

# Question 1
print("\nQuestion 1: What is the capital of France?")
print("A. Berlin")
print("B. Madrid")
print("C. Paris")
print("D. Rome")
answer1 = input("Your answer: ").upper()

if answer1 == 'C':
    print("Correct!")
    score = score + 1
else:
    print(f"Incorrect. The correct answer was C (Paris).")

# Question 2
print("\nQuestion 2: Which planet is known as the Red Planet?")
print("A. Earth")
print("B. Mars")
print("C. Jupiter")
print("D. Venus")
answer2 = input("Your answer: ").upper()

if answer2 == 'B':
    print("Correct!")
    score = score + 1
else:
    print(f"Incorrect. The correct answer was B (Mars).")

# Question 3
print("\nQuestion 3: What is 5 + 7?")
print("A. 10")
print("B. 11")
print("C. 12")
print("D. 13")
answer3 = input("Your answer: ").upper()

if answer3 == 'C':
    print("Correct!")
    score = score + 1
else:
    print(f"Incorrect. The correct answer was C (12).")

# Question 4 (Python related)
print("\nQuestion 4: What keyword is used to define a function in Python?")
print("A. func")
print("B. define")
print("C. def")
print("D. function")
answer4 = input("Your answer: ").upper()

if answer4 == 'C':
    print("Correct!")
    score = score + 1
else:
    print(f"Incorrect. The correct answer was C (def).")

# Question 5 (Python related)
print("\nQuestion 5: Which of these data types is mutable in Python?")
print("A. Tuple")
print("B. String")
print("C. List")
print("D. Integer")
answer5 = input("Your answer: ").upper()

if answer5 == 'C':
    print("Correct!")
    score = score + 1
else:
    print(f"Incorrect. The correct answer was C (List).")

# Question 6 (Python related)
print("\nQuestion 6: How do you start a single-line comment in Python?")
print("A. //")
print("B. /*")
print("C. #")
print("D. <!--")
answer6 = input("Your answer: ").upper()

if answer6 == 'C':
    print("Correct!")
    score = score + 1
else:
    print(f"Incorrect. The correct answer was C (#).")

# Question 7 (Python related)
print("\nQuestion 7: What is the correct way to output 'Hello, World!' in Python?")
print(f"""A. console.log("Hello, World!")""")
print(f"""B. System.out.println("Hello, World!")""")
print(f"""C. print("Hello, World!")""")
# print(f"""D. echo "Hello, World!" """")
answer7 = input("Your answer: ").upper()

if answer7 == 'C':
    print("Correct!")
    score = score + 1
else:
    print(f"Incorrect. The correct answer was C (print(\"Hello, World!\")).")

# Question 8 (Python related)
print("\nQuestion 8: Which operator is used for exponentiation in Python?")
print("A. ^")
print("B. **")
print("C. //")
print("D. %")
answer8 = input("Your answer: ").upper()

if answer8 == 'B':
    print("Correct!")
    score = score + 1
else:
    print(f"Incorrect. The correct answer was B (**).")

# New Question 9 (Python related)
print("\nQuestion 9: What is the output of `type([])` in Python?")
print("A. <class 'tuple'>")
print("B. <class 'list'>")
print("C. <class 'dict'>")
print("D. <class 'str'>")
answer9 = input("Your answer: ").upper()

if answer9 == 'B':
    print("Correct!")
    score = score + 1
else:
    print(f"Incorrect. The correct answer was B (<class 'list'>).")

# New Question 10 (Python related)
print("\nQuestion 10: Which of the following is NOT a valid variable name in Python?")
print("A. my_variable")
print("B. _variable")
print("C. 2variable")
print("D. variable2")
answer10 = input("Your answer: ").upper()

if answer10 == 'C':
    print("Correct!")
    score = score + 1
else:
    print(f"Incorrect. The correct answer was C (2variable).")

print("------------------------------------------")
print(f"Quiz complete! Your final score is {score} out of 10.")

# Performance Analysis (based on 10 questions)
if score >= 8:
    print("Performance: Advanced! Excellent job!")
elif score >= 6:
    print("Performance: Intermediate. Good effort!")
elif score >= 4:
    print("Performance: Beginner. You can do better!")
else:
    print("Performance: Needs improvement. Review the basics.")
```
