# User se number lo:

# agar number > 0 → "Positive"
# warna → "Negative"

```python
number = int(input("Enter the number: "))

if number > 0:
    print("Positive")
elif number == 0:
    print("Zero")
else:
    print("Negative")

print(f""" You enter the number is {number}""")
```

# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
marks = int(input("Enter your marks: "))

if marks >= 90:
  print("Grade A")
elif marks >= 80:
  print("Grade B")
elif marks >= 70:
  print("Grade C")
elif marks >= 60:
  print("Grade D")
else:
  print(f"Your are fail try the next time your marks : {marks}")
```

# User se day number (1–7) lo:

# 1 → Monday
# 2 → Tuesday
## ...
# 7 → Sunday

```python
num = int(input("Enter the 1 to 7 any number: "))

if num == 1:
  print("Monday")
elif num == 2 :
  print("Tuesday")
elif num == 3 :
  print("Wednesday")
elif num == 4 :
  print("Thursday")
elif num == 5 :
  print("Friday")
elif num == 6 :
  print("Saturday")
elif num == 7:
  print("Sunday")
else:
  print(f"You Entered the Wrong Number is {num}")

print(f"You Enterd the number is {num}")
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
month = int(input("Enter a Month number 1 to 12: "))

if month == 12 or month == 1 or month == 2:
  print("Winter")
elif month == 3 or month == 4 or month == 5:
  print("Spring")
elif month == 6 or month == 7 or month == 8:
  print("Summer")
elif month < 1 or month > 12:
  print("Invalid month")
else:
  print("Autumn")

print(f"""You Enter the month num is {month}""")
```

# User se 2 numbers lo:

# pehla > dusra → "First is greater"
# pehla < dusra → "Second is greater"
# warna → "Both equal"

```python
number1 = int(input("Enter first number: "))
number2 = int(input("Enter second number: "))

if number1 > number2:
    print("First is greater")
elif number1 < number2:
    print("Second is greater")
else:
    print("Both equal")

print(f""" You Enter a 2 Number First is {number1} and Second Number is {number2}.""")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
operator = input("Enter the operator (+, -, *, /): ")


if operator == "+":
  print("Addition")
elif operator == "-":
  print("Substraction")
elif operator == "*":
  print("Multiplication")
elif operator == "/":
  print("Division")
else:
  print("Invalid operator")
```
