```python
sub1 = int(input("enter your marks: "))
sub2 = int(input("enter your marks: "))
sub3 = int(input("enter your marks: "))
sub4 = int(input("enter your marks: "))
sub5 = int(input("enter your marks: "))
obtain_marks = sub1+sub2+sub3+sub4+sub5
pec = (obtain_marks/500)*100
grade = ""
result = ""
if sub1>=50:
    if sub2>=50:
        if sub3>=50:
            if sub4>=50:
                if sub5 >=50:
                    result = "Pass "
                else:
                    result = "Fail"
            else:
                result = "Fail"
        else:
            result = "Fail"
    else:
        result = "Fail"
else:
    result = "Fail"

if pec >= 90:
    grade = "A+"
elif pec>=80:
    grade = "B"
elif pec>=70:
    grade = "C"
elif pec>=60:
    grade = "D"
else:
    grade = "Fail"
    
    
print(f"""
      total Marks :    500
      obtain Marks :   {obtain_marks}
      percentage :     {pec}
      Grade     :      {grade} 
      overall result :   {result}
      """)
```
