## ✅ 1. Official List Methods (Total = 11)

These are the *real list methods*:

1. `append()`
2. `extend()`
3. `insert()`
4. `remove()`
5. `pop()`
6. `clear()`
7. `index()`
8. `count()`
9. `sort()`
10. `reverse()`
11. `copy()`

👉 So the correct number is **11 list methods** (not 12)

---

## ⚡ 2. Common Built-in Functions used with lists

These are NOT list methods, but used a lot:

* `len(list)` → size
* `sum(list)` → total
* `max(list)` → largest
* `min(list)` → smallest
* `sorted(list)` → returns sorted copy
* `list()` → convert to list

Example:

```python
a = [3, 1, 5]

len(a)      # 3
max(a)      # 5
sorted(a)   # [1, 3, 5]
```

## 🧠 3. Advanced (from collections module style thinking)

You can also work with lists using:

* `deque` (fast queue)
* `Counter` (count elements)

But these are NOT list methods — they are separate tools.
