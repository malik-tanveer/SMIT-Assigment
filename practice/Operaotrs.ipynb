## Python Operators

Operators are special symbols that perform operations on values and variables. Python provides several types of operators:

1.  **Arithmetic Operators**: Perform mathematical operations.
2.  **Comparison (Relational) Operators**: Compare two values and return a boolean result.
3.  **Assignment Operators**: Assign values to variables.
4.  **Logical Operators**: Combine conditional statements.
5.  **Identity Operators**: Compare the memory locations of two objects.
6.  **Membership Operators**: Test if a sequence contains a specified value.
7.  **Bitwise Operators**: Perform operations on individual bits of numbers.

#Arithmetic Operators


# +, - , * , / , % , // , **



```python
a = 10
b = 3

print(f"a + b = {a + b}")
print(f"a - b = {a - b}")
print(f"a * b = {a * b}")
print(f"a / b = {a / b}")
print(f"a % b = {a % b}")
print(f"a // b = {a // b}")
print(f"a ** b = {a ** b}")
```

# Comparison (Relational) Operators

# == , != , > , < , >= , <=

```python
x = 15
y = 10

print(f"x == y: {x == y}")
print(f"x != y: {x != y}")
print(f"x > y: {x > y}")
print(f"x < y: {x < y}")
print(f"x >= y: {x >= y}")
print(f"x <= y: {x <= y}")
```

# Assignment Operators

# += , -= , *= , /=

```python
c = 20
print(f"Initial c = {c}")
c += 5
print(f"c += 5: {c}")
c -= 3
print(f"c -= 3: {c}")
c *= 2
print(f"c *= 2: {c}")
c /= 4
print(f"c /= 4: {c}")
```

# Logical Operators

# and , or , not

```python
p = True
q = False

print(f"p and q: {p and q}")
print(f"p or q: {p or q}")
print(f"not p: {not p}")
```

# Identity Operators

```python
a = 10
b = 10

name1 = "Rehman"
name2 = "Khan"


print(f"{a} and {b} is Equal Output True: {a is b}")
print(f"{a} and {b} is Equal Output False: {a is not b}")


print(f"{name1} and {name1} is Equal Output True: {name1 is name1}")
print(f"{name1} and {name2} is Equal Output False: {name1 is name2}")
print(f"{name2} and {name2} is Equal Output True: {name2 is name2}")
```
