#Tuple in Python

Tuples are ordered, immutable collections of items. They are similar to lists but cannot be modified after creation. Tuples are often used to store heterogeneous data (different data types) and are typically created with parentheses `()`. Due to their immutability, they can be used as keys in dictionaries and elements in sets.

```python
# This is a string not a tuple you write a like and then tuple "Hello", this a Tuple
a = ("hello") #
type(a)
```
